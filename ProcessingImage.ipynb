```python
from skimage import color, io
import numpy as np
from IPython.display import Image
```

```python
img = io.imread("imagens/DKC.jpg")
io.imshow(img)
io.show()
```

```python
array_image = np.asarray(img)
print("Representao Numrica da Imagem\n\n %r" %array_image)
```

```python
gray_image = color.rgb2gray(img)
io.imshow(gray_image)
io.show()
```

```python
io.imsave("imagens/DKC_grayscale.jpg", gray_image)
```

```python
hsv_image = color.rgb2hsv(img)
io.imshow(hsv_image)
io.show()
```

```python
io.imsave("imagens/DKC_hsv.jpg", hsv_image)
```

```python
lab_image = color.rgb2lab(img)
io.imshow(lab_image)
io.show()
```

```python
io.imsave("imagens/DKC_lab.jpg", lab_image)
```

```python
luv_image = color.rgb2luv(img)
io.imshow(luv_image)
io.show()
```
