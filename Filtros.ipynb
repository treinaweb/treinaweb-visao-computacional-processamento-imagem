```python
from skimage import morphology, filters, io, color, feature
```

```python
img = io.imread("imagens/DKC2.jpg")
io.imshow(img)
io.show()
```

# Filtro Erosão

```python
img_erosao = morphology.erosion(img)
io.imshow(img_erosao)
io.show()
```

# Filtro Dilatação

```python
img_dilatacao = morphology.dilation(img)
io.imshow(img_dilatacao)
io.show()
```

# Filtro Sobel

```python
img_cinza = color.rgb2gray(img)
io.imshow(img_cinza)
io.show()
```

```python
sobel_img = filters.sobel(img_cinza)
io.imshow(sobel_img)
io.show()
```

# Filtro Canny

```python
canny_img = feature.canny(img_cinza, 3)
io.imshow(canny_img)
io.show()
```
